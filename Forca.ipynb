```python
import random
import requests
import unicodedata

#Cria a lista de palavras aleatórias a partir do Dicionário Aberto
def remover_acentos(texto):
    """Remove acentos e deixa apenas letras (a-z)."""
    nfkd = unicodedata.normalize('NFKD', texto)
    return ''.join([c for c in nfkd if c.isalpha()])

def gerar_palavra_aleatoria():
    """Busca uma palavra real aleatória do Dicionário Aberto."""
    url = "https://api.dicionario-aberto.net/random"
    try:
        resposta = requests.get(url, timeout=5)
        if resposta.status_code == 200:
            dados = resposta.json()
            #print(dados)
            palavra = dados.get("word")

            if palavra:
                palavra = remover_acentos(palavra.lower())

                # filtra palavras curtas, longas ou esquisitas
                if 4 <= len(palavra) <= 10:
                    return palavra
    except requests.exceptions.RequestException:
        print("⚠️ Erro ao conectar à API.")

    return None


# Gera uma lista de 10 palavras aleatórias únicas
palavras = set()
while len(palavras) < 10:
    palavra = gerar_palavra_aleatoria()
    if palavra:
        palavras.add(palavra)

#print("Palavras geradas:", list(palavras))


#escolhe uma palavra aleatória da lista
palavra_sorteada = random.choice(list(palavras))

#Criamos uma string com traços que representao as letras
palavra_oculta = "*" * len(palavra_sorteada)

#Criamos uma lista vazia para armazenar as letras já tentadas
letras_tentadas = []

max_tentativas = 6
tentativas_restantes = max_tentativas

while True:
    #Mostra a palavra oculta
    print("Palavra:", palavra_oculta)

    #Pedimos para o jogador digitar uma letra
    letra = input("Digite uma letra: ").lower()

    #Verificamos se a letra ja foi digitada
    if letra in letras_tentadas: 
        print("Você já tentou essa letra. Tente outra.")
        continue

    #Adicionamos a letra na lista de letras tentadas
    letras_tentadas.append(letra)

    #Verificamos se a letra está na palavra sorteada
    if letra in palavra_sorteada:
        #Atualizamos a palavra oculta
        lista = []
        for i in range(len(palavra_sorteada)):      #Percorre cada indice da palavra sorteada
            if palavra_sorteada[i] == letra:        #Se a letra na posicao i for igual a letra digitada
                lista.append(letra)                 #Adiciona a letra na lista
            else:
                lista.append(palavra_oculta[i])     #Se não for, Adiciona o caractere atual da palavra oculta na lista
        palavra_oculta = "".join(lista)             #Atualiza a palavra oculta com a nova lista

    else: #Letra incorreta
        tentativas_restantes -= 1
        print(f"Letra incorreta. Você tem {tentativas_restantes} tentativas restantes.")

    #Verificamos se o jogador ganhou
    if palavra_oculta == palavra_sorteada:
        print("Parabéns! Você ganhou! A palavra era:", palavra_sorteada)
        break
    elif tentativas_restantes == 0:
        print("Você perdeu! A palavra era:", palavra_sorteada)
        break
```
